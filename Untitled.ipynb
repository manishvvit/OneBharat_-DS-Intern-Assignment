```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# Load the bank statement JSON data
with open('P1- BankStatements.json', 'r') as file:
    data = json.load(file)

# Extract transaction data
transactions = data['Account']['Transactions']['Transaction']

# Create a DataFrame from the transaction data
df = pd.DataFrame(transactions)

# Convert necessary columns to appropriate data types
df['transactionTimestamp'] = pd.to_datetime(df['transactionTimestamp'])
df['amount'] = pd.to_numeric(df['amount'])
df['currentBalance'] = pd.to_numeric(df['currentBalance'])

# Display the first few rows of the DataFrame
print("First few transactions:")
print(df.head())

# Calculate balance trends over time
plt.figure(figsize=(12, 6))
plt.plot(df['transactionTimestamp'], df['currentBalance'], marker='o')
plt.title('Balance Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Balance (INR)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Analyze spending patterns
debit_transactions = df[df['type'] == 'DEBIT']
spending_by_mode = debit_transactions.groupby('mode')['amount'].sum()

# Plot spending patterns
plt.figure(figsize=(10, 6))
spending_by_mode.plot(kind='bar')
plt.title('Spending Patterns by Mode')
plt.xlabel('Mode')
plt.ylabel('Total Amount Spent (INR)')
plt.grid(axis='y')
plt.show()

# Income analysis
credit_transactions = df[df['type'] == 'CREDIT']
income_by_mode = credit_transactions.groupby('mode')['amount'].sum()

# Plot income sources
plt.figure(figsize=(10, 6))
income_by_mode.plot(kind='bar')
plt.title('Income Sources by Mode')
plt.xlabel('Mode')
plt.ylabel('Total Income (INR)')
plt.grid(axis='y')
plt.show()

# Detect anomalies (e.g., large transactions)
large_transactions = df[df['amount'] > 500]  # Define a threshold for "large"
print("Large transactions (amount > 500 INR):")
print(large_transactions)

# Generate alerts for potential anomalies (transactions > 2000 INR)
alerts = df[df['amount'] > 2000]
print("Alerts for transactions exceeding 2000 INR:")
print(alerts)
```

```python
# Define criteria for small and large transactions
small_threshold = 500  # Transactions <= 500 are considered small
large_threshold = 2000 # Transactions > 2000 are considered large

# Categorize transactions
df['transaction_size'] = df['amount'].apply(lambda x: 'Small' if x <= small_threshold else ('Large' if x > large_threshold else 'Medium'))

# Count the number of transactions in each category
transaction_size_distribution = df['transaction_size'].value_counts()
print("Distribution of Transaction Amounts:")
print(transaction_size_distribution)
```

```python
# Extract transaction data
transactions = data['Account']['Transactions']['Transaction']

# Create a DataFrame from the transaction data
df = pd.DataFrame(transactions)

# Convert necessary columns to appropriate data types
df['transactionTimestamp'] = pd.to_datetime(df['transactionTimestamp'])
df['amount'] = pd.to_numeric(df['amount'])
df['currentBalance'] = pd.to_numeric(df['currentBalance'])

# Sort the DataFrame by transactionTimestamp to ensure chronological order
df = df.sort_values(by='transactionTimestamp')

# Task 1: Plot the Account Balance Trend Over Time
plt.figure(figsize=(12, 6))
plt.plot(df['transactionTimestamp'], df['currentBalance'], marker='o', linestyle='-', color='blue')
plt.title('Account Balance Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Account Balance (INR)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Task 2: Identify Significant Balance Changes
# Calculate the daily change in balance
df['balance_change'] = df['currentBalance'].diff()

# Define a threshold for significant changes (e.g., changes greater than a certain percentage of the balance)
significant_change_threshold = df['currentBalance'].mean() * 0.1  # Change greater than 10% of the mean balance

# Find periods with significant balance changes
significant_changes = df[abs(df['balance_change']) > significant_change_threshold]

# Display significant balance changes
print("Significant Balance Changes (greater than 10% of the mean balance):")
print(significant_changes[['transactionTimestamp', 'currentBalance', 'balance_change']])

# Plot the significant balance changes on the balance trend plot
plt.figure(figsize=(12, 6))
plt.plot(df['transactionTimestamp'], df['currentBalance'], marker='o', linestyle='-', color='blue', label='Balance')
plt.scatter(significant_changes['transactionTimestamp'], significant_changes['currentBalance'], color='red', label='Significant Change', zorder=5)
plt.title('Account Balance Trend with Significant Changes Highlighted')
plt.xlabel('Date')
plt.ylabel('Account Balance (INR)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_json('P1- BankStatements.json')

# Example descriptions for categorization; adapt these rules to your specific data
def categorize_transaction(description):
    description = description.lower()
    if 'fuel' in description:
        return 'Fuel'
    elif 'ecommerce' in description or 'shopping' in description:
        return 'E-commerce/Shopping'
    elif 'food' in description or 'restaurant' in description:
        return 'Food'
    elif 'atm' in description:
        return 'ATM Withdrawals'
    elif 'upi' in description:
        return 'UPI Transactions'
    else:
        return 'Other'

# Apply the categorization function
df['category'] = df['description'].apply(categorize_transaction)

# Analyze the frequency of transactions in each category
category_frequency = df['category'].value_counts()

# Analyze the total amount of spending in each category
category_amount = df.groupby('category')['amount'].sum()

# Plot the frequency of transactions in each category
plt.figure(figsize=(12, 6))
category_frequency.plot(kind='bar', color='skyblue')
plt.title('Frequency of Transactions in Each Category')
plt.xlabel('Category')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.savefig('/mnt/data/frequency_of_transactions.png')  # Save the plot as an image
plt.show()

# Plot the total amount of spending in each category
plt.figure(figsize=(12, 6))
category_amount.plot(kind='bar', color='orange')
plt.title('Total Amount Spent in Each Category')
plt.xlabel('Category')
plt.ylabel('Total Amount (INR)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.savefig('/mnt/data/total_amount_spent.png')  # Save the plot as an image
plt.show()

# Display the analyzed data
print("Frequency of Transactions in Each Category:")
print(category_frequency)
print("\nTotal Amount Spent in Each Category:")
print(category_amount)
```
